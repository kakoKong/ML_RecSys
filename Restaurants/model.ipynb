```python
import pandas as pd
```

```python
restaurants = pd.read_csv('./data/main_data.csv', engine='python');
restuarants = restaurants.drop(columns = 'Unnamed: 0')
restuarants
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df = 3,  max_features = None, strip_accents='unicode', 
                          analyzer='word', ngram_range=(1, 3),
                          stop_words = 'english'
                         )
restaurants['Reviews'] = restaurants['Reviews'].fillna('')
restaurants['Cuisine Style'] = restaurants['Cuisine Style'].fillna('')
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
#For review
X = tfv.fit_transform(restuarants['Reviews'])
Y = tfv.fit_transform(restuarants['Cuisine Style'])
```

```python
sig = sigmoid_kernel(X, X)
```

```python
sig2 = sigmoid_kernel(Y, Y)
```

```python
sig[0]
indices = pd.Series(restaurants.index, index=restaurants['Name']).drop_duplicates()
indices
```

```python
def give_rec(title, sig=sig2):
    idx = indices[title]
    
    sig_scores = list(enumerate(sig[idx]))
    
#     sig_scores2 = list(enumerate(sig2[idx]))
    
#     print(sig_scores1[1].apply(lambda x: x*0.6))
    
#     sig_scores = (sig_scores2[0] * 0.6) + (sig_scores1[0] * 0.4)
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
    sig_scores = sig_scores[1:11]
    
    res_indices = [i[0] for i in sig_scores]
    
    return restuarants['Name'].iloc[res_indices]
```

```python
give_rec('Il Padrino')
```
